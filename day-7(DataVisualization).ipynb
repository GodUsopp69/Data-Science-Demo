```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.0,5.0,100)
y = np.sin(x)

# print(x)
# print(y)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot Ex')
plt.grid(True)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,5,7,2,1])

plt.scatter(x,y, c='red', marker='o', edgecolors='black')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot Ex')
plt.grid(True)
plt.show()
```

```python
cat = ['A','B','C','D']
valu = [20,35,10,15]


plt.bar(cat,valu, color = ['red','green','blue','pink'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Example')
plt.xticks(rotation=180)
plt.show()
```
